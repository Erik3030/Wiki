```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_bytes(item):
    if len(item.findAll("span", {"class":"mw-plusminus-pos mw-diff-bytes"})) == 1:
        return int(item.findAll("span", {"class":"mw-plusminus-pos mw-diff-bytes"})[0].text)
    elif len(item.findAll("span", {"class":"mw-plusminus-neg mw-diff-bytes"})) == 1:
        return int(item.findAll("span", {"class":"mw-plusminus-neg mw-diff-bytes"})[0].text)
    elif len(item.findAll("strong", {"class":"mw-plusminus-neg mw-diff-bytes"})) == 1:
        return int(item.findAll("strong", {"class":"mw-plusminus-neg mw-diff-bytes"})[0].text.replace(",", ""))
    elif len(item.findAll("strong", {"class":"mw-plusminus-pos mw-diff-bytes"})) == 1:
        return int(item.findAll("strong", {"class":"mw-plusminus-pos mw-diff-bytes"})[0].text.replace(",", ""))
    else:
        return 0
    
def get_page(item):
    return item.findAll("span", {"class":"mw-changeslist-line-inner"})[0]["data-target-page"]

def get_url(item):
    try:
        return "https://en.wikipedia.org" + item.findAll("span", {"class":"mw-title"})[0].findAll("a")[0]["href"]
    except:
        return None
    
def get_diff_url(item):
    try:
        return "https://en.wikipedia.org" + item.findAll("div",
        {"class":"mw-changeslist-links"})[0].findAll("a")[0]["href"]
    except:
        return None
```

```python
resp = requests.get("https://en.wikipedia.org/wiki/Special:RecentChanges?hidebots=1&hidecategorization=1&hideWikibase=1&limit=50&days=7&urlversion=2")
soup = BeautifulSoup(resp.content, "html.parser")

rows = []

for item in soup.findAll("ul", {"class":"special"})[0].findAll("li"):
    byte = get_bytes(item)
    title = get_page(item)
    url = get_url(item)
    diff_url = get_diff_url(item)
    rows.append({
        "Bytes":byte,
        "Title":title,
        "Url":url,
        "Diff_Url":diff_url
    })
    
pd.DataFrame(rows)
```
